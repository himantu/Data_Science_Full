```python
match_series = df['match_winner'].value_counts()
sns.barplot(y= match_series.index, x= match_series.values)
```

```python
top = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
top
```

```python
top.plot(kind= 'bar')
```

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x : x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
df.head()
```

```python
topb = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
topb.plot(kind='bar')
```
